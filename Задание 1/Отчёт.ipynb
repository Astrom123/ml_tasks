### 1. В чем заключается идея использования CNN для классификации текстов и какие у этого подхода есть особенности?

CNN для классификации текстов применяется следующим образом:
1. Все слова представляются в виде векторов размера k.
2. Каждое предложение представляется в виде матрицы размера n x k.
3. Для получения признаков из матрицы используются фильтры и операция свёртки.


Фильтр - это матрица размера h x k, где h - размер окна. При применении операции свёртки к фильтру и каждому возможному окну в предложении получается карта признаков. Из всех признаков выбирается максимум (наиболее важный признак).

### 2. На каких наборах данных проводилось сравнение моделей?

* MR - отзывы о фильмах. Классификация: негативный, положительный
* SST-1 - MR с более детальной классификацией. Классификация: очень негативный, ..., нейтральный, ..., очень положительный
* SST-2 - SST-1 без класса 'нейтральный'.
* Subj - суждения. Классификация: субъективное, объективное
* TREC - вопросы. Классификация: о человеке, о месте, о числе, ...
* CR - отзывы покупателей о товарах. Классификация: негативный, положительный
* MPQA - новостные статьи. Классификация: положительный, отрицательный, нейтральный, скорее положительный, ...

### 3. Использовалась ли предобработка данных (например, лемматизация)?

Для превращения слов в вектора использовалась модель world2vec, обученная на большом количестве данных.

### 4. В экспериментах использовалось несколько типов моделей, основанных на CNN. В чем их отличия?

* CNN-rand - все слова инициализируются случайными векторами, а в процессе обучения модифицируются.
* CNN-static - слова в векторном виде представляются с помощью world2vec, и в процессе обучения не модифицируются. Неизвестные слова инициализируются случайным образом.
* CNN-non-static - слова в векторном виде представляются с помощью world2vec, и в процессе обучения настраиваются.
* CNN-multichannel - каждое предложение представляется двумя каналами. Слова в обоих каналах инициализируются с помощью world2vec. Фильтры применяются к обоим каналам, но настраивается в процессе обучения только один канал.

### 5. К каким выводам можно прийти после прочтения статьи?

По результатам сравнения CNN моделей и других алогритмов видно, что CNN не уступают другим известным алгоритмам в NLP задачах.

По результатам среди расмматриваемых CNN моделей можно сделать слеюдущие выводы:
* CNN-rand показала результаты значительно хуже относительно остальных моделей. Использование предварительно подготовленных моделей для перевода слов в вектора даёт лучший результат.
* В некоторых задачах CNN-non-static даёт прирост к результату. Обучение позволяет адаптироваться натренированным словам к необходимой задаче и установить связи между не известными ранее словами.
